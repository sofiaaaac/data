```python
colunas_sim_nao = ['FEBRE','MIALGIA','CEFALEIA',
            'EXANTEMA','VOMITO','NAUSEA','DOR_COSTAS',
            'CONJUNTVIT','ARTRITE','ARTRALGIA','PETEQUIA_N',
            'LEUCOPENIA','LACO','DOR_RETRO','DIABETES',
            'HEMATOLOG','HEPATOPAT','RENAL','HIPERTENSA',
            'ACIDO_PEPT','AUTO_IMUNE', 'HOSPITALIZ',
            'ALRM_HIPOT','ALRM_PLAQ','ALRM_VOM','ALRM_SANG',
            'ALRM_HEMAT','ALRM_ABDOM','ALRM_LETAR','ALRM_HEPAT',
            'ALRM_LIQ','DT_ALRM','GRAV_PULSO','GRAV_CONV',
            'GRAV_ENCH','GRAV_INSUF','GRAV_TAQUI','GRAV_EXTRE',
            'GRAV_HIPOT','GRAV_HEMAT','GRAV_MELEN','GRAV_METRO',
            'GRAV_SANG','GRAV_AST','GRAV_MIOC','GRAV_CONSC']


dicionario_colunas = {
    'TPAUTOCTO':{1:'Sim',
                 2:'No',
                 3:'Indeterminado'},
    
    'CLASSI_FIN':{5:'Descartado',
                 10:'Dengue',
                 11:'Dengue com sinais de alarme',
                 12:'Dengue grave',
                 13:'Chikungunya',
                 8:'Fechado pelo sistema'},
    
    'EVOLUCAO':{1:'Cura',
               2:'bito pelo agravo',
               3:'bito por outras causas',
               4:'bito em investigao',
               9:'Ignorado'},
    
    'SOROTIPO':{1:'DEN1',
               2:'DEN2',
               3:'DEN3',
               4:'DEN4',},
    
    'CS_RACA':{ 1: 'Branca',
               2: 'Preta',
               3: 'Amarela',
               4: 'Parda',
               5: 'Indgena',
               9: 'Ignorado' },
    
    'CS_ESCOL_N':{ 0: 'Analfabeto',
                  1: '1 a 4 srie incompleta do EF',
                  2: '4 srie completa do EF ( antigo 1 grau)',
                  3: '5  8 srie incompleta do EF (antigo ginsio ou 1 grau)',
                  4: 'Ensino fundamental completo (antigo ginsio ou 1 grau)',
                  5: 'Ensino mdio incompleto (antigo colegial ou 2 grau)',
                  6: 'Ensino mdio completo (antigo colegial ou 2 grau)',
                  7: 'Educao superior incompleta',
                  8: 'Educao superior completa',
                  9: 'Ignorado',
                  10: 'No se aplica' },
    
    'CS_GESTANT':{ 1: '1 Trimestre',
                  2: '2 Trimestre',
                  3: '3 Trimestre',
                  4: 'Idade gestacional ignorada',
                  5: 'No',
                  6: 'No se aplica',
                  9: 'Ignorado'},
    
    'CS_ZONA':{1:'Urbana',
                        2:'Rural',
                        3:'Periurbana',
                        9:'Ignorado'}
}

for coluna in dicionario_colunas.keys():
    dados[coluna] = dados[coluna].replace(dicionario_colunas[coluna])

for i in colunas_sim_nao:
    dados[i] = dados[i].replace({1:'Sim',
                                2:'No',
                                3:'Indeterminado'})
    
dados.head()
```

```python
municipios = pd.read_csv('Municpios RS IBGE6 Populao CRS Regional.csv')
municipios = municipios[['IBGE6','Municpio','Populao_estimada','CRS']]
municipios = municipios.set_index('IBGE6')
dicionario_municipios = municipios.to_dict()['Municpio']
dicionario_CRS = municipios.to_dict()['CRS']

dados['ID_MN_RESI_nome'] = dados['ID_MN_RESI'].map(dicionario_municipios).fillna('Fora do RS')
dados['ID_MUNICIP_nome'] = dados['ID_MUNICIP'].map(dicionario_municipios).fillna('Fora do RS')
dados['COMUNINF_nome'] = dados['COMUNINF'].map(dicionario_municipios)


dados['CRS'] = dados['ID_MN_RESI'].map(dicionario_CRS)

#dados = dados.rename(columns={'ID_MN_RESI':,
#                              'SOROTIPO':,
#                              'HOSPITALIZ':,
#                              'TPAUTOCTO':,
#                              'CLASSI_FIN':,
#                              'EVOLUCAO':,
#                              'ID_MN_RESI_nome':,
#                              'CRS':})

dados.head()
```
